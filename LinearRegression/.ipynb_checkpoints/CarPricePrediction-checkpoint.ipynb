```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_csv(r"Car details v3.csv")
```

Checking for Null Values

```python
car.isnull().sum()
```

```python
car[car["mileage"].isnull()]
```

Since values are null at same positions, it can be a option to drop the that rows from the dataset

```python
car=car.dropna()
```

```python
car.shape
```

```python
#torque column is categorical hence dropping it
car.drop(['torque'],1,inplace=True)
```

```python

plt.figure(figsize=(12,8))
car.name.value_counts().head(10).plot(kind='bar')
plt.title("The top 10 selling cars")
plt.show()
```

```python
car.drop(["name"],axis=1,inplace=True)
```

Conersion of object into float for better analysis

```python
car['mileage']=car.mileage.apply(lambda x:float(x.split(' ')[0]))
car['engine']=car.engine.apply(lambda x:int(x.split(' ')[0]))
car['max_power']=car.max_power.apply(lambda x:float(x.split(' ')[0]))
```

```python
car["years_driven"]=2021-car["year"]
```

```python
car.drop(["year"],axis=1,inplace=True)
```

# Understanding the Data

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
sns.countplot(car["fuel"])
plt.subplot(2,2,2)
sns.countplot(car["seller_type"])
plt.subplot(2,2,3)
sns.countplot(car["transmission"])
plt.subplot(2,2,4)
sns.countplot(car["owner"])
plt.show()
```

```python
sns.pairplot(car)
```

```python
car.drop(["fuel","transmission","owner","seller_type"],axis=1,inplace=True)
```

# Checking for good predictors

```python
plt.figure(figsize=(16,10))
sns.heatmap(car.corr(),annot=True,cmap="Blues")
plt.show()
```

```python
car.drop("seats",axis=1,inplace=True)
```

# Removing Outliers

```python
plt.figure(figsize=(15,5))
plt.subplot(2,2,1)
sns.boxplot(car["max_power"])
plt.subplot(2,2,2)
sns.boxplot(car["engine"])
plt.subplot(2,2,3)
sns.boxplot(car["mileage"])
```

```python
car=car[(car["max_power"]<260)]
car=car[car["engine"]<3000]
car=car[(car["mileage"]<32)|(car["mileage"]>5)]
```

```python
x=car.drop(['selling_price'],1)
y=car['selling_price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
y_pred = regr.predict(x_test)
plt.scatter(x_test["mileage"], y_test, color ='b')
plt.scatter(x_test["mileage"], y_pred, color ='k')
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print("The R^2 score for this model is : ",r2)
```
